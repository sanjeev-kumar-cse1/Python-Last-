```python
# You are building a college grading system. Write a function that takes students marks and return on the basis of marks (out of 100)
# 90-100:A
# 80-90:B
# 70-80:C
# 60-70:D
# 50-0:Fail

def students(marks_list):
    for marks in marks_list:
        if 90 <= marks <= 100:
            print("A")
        elif 80 <= marks < 90:
            print("B")
        elif 70 <= marks < 80:
            print("C")
        elif 60 <= marks < 70:
            print("D")
        else:
            print("Fail")
    print(students(50))
```

```python
# Write a function to reverse the words of sentence 

def reverse_words(sentence):
    words = sentence.split()
    reversed_words = words[::-1]
    return ' '.join(reversed_words)
sentence = "Hello world, Python is fun"
reversed_sentence = reverse_words(sentence)
print(reversed_sentence)
```

```python
# Write a function to reverse the words of sentence 

sent = "Hello world, Python is fun".split()
rev = []
for word in sent:
    rev.append(word[::-1])
reversed_sent = ' '.join(rev)
print(reversed_sent)

```

```python
# Check the two strings are "Anagram" or not. Like all letters are match and create new words from that all letter of first "silent = listen"

def are_anagrams(str1, str2):
    str1 = str1.replace(" ", "").lower()
    str2 = str2.replace(" ", "").lower()
    return sorted(str1) == sorted(str2)
word1 = "silent"
word2 = "listen"
if are_anagrams(word1, word2):
    print(f'"{word1}" and "{word2}" are anagrams.')
else:
    print(f'"{word1}" and "{word2}" are not anagrams.')
```

```python
# Other method

li = [1, 2, 3, 4, 5]
left = 0
right = len(li) - 1
while left < right:
    li[left], li[right] = li[right], li[left]
    left += 1
    right -= 1
print(li)
```
